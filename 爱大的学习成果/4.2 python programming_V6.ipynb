## Pre-processing Text Data

In order to capture the meaningful sentiment, for both positive and negative, the text data cleaning is executed, by using re.sub() method to remove any irrelevant punctuations and tags characters

### Remove Punctuations and Tags

```python
# Show all the punctuation characters
string.punctuation
```

```python
# Define a function to remove all the punctuation characters and numerical values that convey no meaning from the texts 
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

# Define a function to remove all the html tags that convey no meaning from the texts 
remove_html = re.compile('<.*?>')

def remove_tags(text):
    return re.sub(remove_html, '', text).strip()

# Create the Clean Tweet Dataframe after punctuations and tags removal
df['Clean Tweet'] = df['Processed tweet'].apply(lambda x: remove_punct(x),remove_tags(x))
df.drop(['Processed tweet'], axis=1, inplace=True)
df.head(10)
```

##  Stemming and Lemmitization

>Stemming applies algorithmic rules to extract the stem out of the derived word simply. The words produced by this step do not have any essential meaning, but they are simply a bunch of letters put together without affixes. For example, the word “beautiful” is stemmed to “Beauti”.

>Lemmatization is similar to stemming but it add context to bring out the true meaning. It groups inflected forms of words to be interpreted as a single root word. For example, the word “beautiful” is stemmed to “Beauty” unlike “Beauti”.

```python
# Define the function to extract the stem from the derived word in the texts
ps = nltk.PorterStemmer()

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

df['Clean Tweet'] = df['Clean Tweet'].apply(lambda x: stemming(x))
df.head()
```

```python
# Define the function to help bring the true meaning of the words
def basic_clean(text):
    """
    A simple function to clean up the data. All the words that are not designated 
    as a stopword is then lemmatized after encoding and basic regex parsing are performed
    """
    wnl = nltk.stem.WordNetLemmatizer()
    stopwords = nltk.corpus.stopwords.words('english')
    words = re.sub(r'[^\w\s]', '', text).split()
    return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

# Building a Basic Model

```python
# Run the logistic regression model and prediction of Clean Tweet with regards to sentiment
# Return the accuracy rate
x_train,x_test,y_train,y_test = train_test_split(df_visulization['Clean Tweet'], df_visulization['sentiment'], test_size=0.2, random_state=2020)

pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', LogisticRegression())])

model = pipe.fit(x_train, y_train)
prediction = model.predict(x_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, prediction)*100,2)))
```

```python
# Show the confusion matrix
print(confusion_matrix(y_test, prediction))
print(classification_report(y_test, prediction))
```

```python
# Plot the confusion matrix in the heatmap
plot_confusion_matrix(model,x_test,y_test)
```

```python
# Compute precision-recall pairs for different probability thresholds

# Compute Receiver operating characteristic (ROC)
fpr, tpr, thresholds = roc(y_test,prediction)
# Compute Area Under the Curve (AUC) using the trapezoidal rule.
roc_auc = auc(fpr,tpr)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.plot(fpr, tpr, color='darkorange',
         lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.title('Receiver operating characteristics curve')
plt.legend(loc="lower right")
plt.show()
```

### Observations from a Basic Model
- By applying logistic regression that suits sentiment value (0 for negative and  1 for positive), we can observe the accuracy of the prediction model
